```python
# LIBRARIES
import os
import json
import pandas as pd
import pymysql
import geopandas
```

```python
# AGGREGATE TRANSACTIONS

path_agg_trans = "C:/Users/muthu/Desktop/guvi project/capstone 2 Phonepe/pulse/data/aggregated/transaction/country/india/state"
Agg_trans_state_list = os.listdir(path_agg_trans)
# Agg_trans_state_list

columns_agg_trans={'States':[], 'Year':[],'Quarter':[],'Transaction_Type':[], 'Transaction_Count':[], 'Transaction_Amount':[]}
for i_state in Agg_trans_state_list:
    states = path_agg_trans+"/"+i_state+"/"
    # print(states)
    Agg_trans_year = os.listdir(states)
    
    for j_year in Agg_trans_year:
        Agg_T_year_list=states+j_year+"/"
        Agg_year_details=os.listdir(Agg_T_year_list)
        
        for k_year in Agg_year_details:
            inside_Year=Agg_T_year_list+k_year
            data=open(inside_Year,'r')
            year_data=json.load(data)
            
            for z_year in year_data['data']['transactionData']:
                Name=z_year['name']
                count=z_year['paymentInstruments'][0]['count']
                amount=z_year['paymentInstruments'][0]['amount']
                columns_agg_trans['Transaction_Type'].append(Name)
                columns_agg_trans['Transaction_Count'].append(count)
                columns_agg_trans['Transaction_Amount'].append(amount)
                columns_agg_trans['States'].append(i_state)
                columns_agg_trans['Year'].append(j_year)
                columns_agg_trans['Quarter'].append(int(k_year.strip('.json')))

Aggregated_Trans = pd.DataFrame(columns_agg_trans) 
Aggregated_Trans['States']=Aggregated_Trans['States'].str.title() 
Aggregated_Trans['States']=Aggregated_Trans['States'].str.replace("-"," ")              
# Aggregated_Trans 
```

```python
# AGGREGATED USERS

path_agg_users = "C:/Users/muthu/Desktop/guvi project/capstone 2 Phonepe/pulse/data/aggregated/user/country/india/state"
Agg_user_state_list = os.listdir(path_agg_users)
# Agg_user_state_list

columns_agg_users={'States':[], 'Year':[],'Quarter':[],'Brands':[], 'Count':[], 'Percentage':[]}
for i_state in Agg_user_state_list:
    states=path_agg_users+"/"+i_state+"/"
    Agg_users_year=os.listdir(states)
    
    for j_year in Agg_users_year:
        Agg_U_year_list=states+j_year+"/"
        Agg_year_users=os.listdir(Agg_U_year_list)
        
        for k_year in Agg_year_users:
            inside_Year=Agg_U_year_list+k_year
            data=open(inside_Year,'r')
            year_data=json.load(data)
            
            try:
                for z_year in year_data['data']['usersByDevice']:
                    Brand=z_year['brand']
                    Count=z_year['count']
                    Percentage=z_year['percentage']
                    columns_agg_users['Brands'].append(Brand)
                    columns_agg_users['Count'].append(Count)
                    columns_agg_users['Percentage'].append(Percentage)
                    columns_agg_users['States'].append(i_state)
                    columns_agg_users['Year'].append(j_year)
                    columns_agg_users['Quarter'].append(int(k_year.strip('.json')))
            except:
                pass
Aggregated_Users=pd.DataFrame(columns_agg_users)
Aggregated_Users['States']=Aggregated_Users['States'].str.title() 
Aggregated_Users['States']=Aggregated_Users['States'].str.replace("-"," ")       
# Aggregated_Users               
```

```python
# MAP TRANSACTIONS

path_map_trans = "C:/Users/muthu/Desktop/guvi project/capstone 2 Phonepe/pulse/data/map/transaction/hover/country/india/state"
map_trans_state_list = os.listdir(path_map_trans)
# map_trans_state_list

columns_map_trans={'States':[], 'Year':[],'Quarter':[],'District_Name':[], 'Transaction_Count':[], 'Transaction_Amount':[]}
for i_state in map_trans_state_list:
    states = path_map_trans+"/"+i_state+"/"
    # print(states)
    map_trans_year = os.listdir(states)
    
    for j_year in map_trans_year:
        map_T_year_list=states+j_year+"/"
        map_year_details=os.listdir(map_T_year_list)
        
        for k_year in map_year_details:
            inside_Year=map_T_year_list+k_year
            data=open(inside_Year,'r')
            year_data=json.load(data)
            
            for z_year in year_data['data']['hoverDataList']:
                Name=z_year['name']
                count=z_year['metric'][0]['count']
                amount=z_year['metric'][0]['amount']
                columns_map_trans['District_Name'].append(Name)
                columns_map_trans['Transaction_Count'].append(count)
                columns_map_trans['Transaction_Amount'].append(amount)
                columns_map_trans['States'].append(i_state)
                columns_map_trans['Year'].append(j_year)
                columns_map_trans['Quarter'].append(int(k_year.strip('.json')))
Map_Transactions= pd.DataFrame(columns_map_trans)
Map_Transactions['States']=Map_Transactions['States'].str.title() 
Map_Transactions['States']=Map_Transactions['States'].str.replace("-"," ") 
# Map_Transactions
```
